```python
import pandas as pd
```

```python
df = pd.read_csv("./data/raw/gdp_per_capita.csv")
regions = pd.read_csv("./data/reference/wb_regions.csv")
df.head()
```

```python
regions.rename(columns={"World Region according to the World Bank": "Region"}, inplace=True)
```

```python
df_merged = pd.merge(left=df, right=regions, how="inner", left_on="Code", right_on="Code", suffixes=["","_r"])
df_merged = df_merged.drop(['Entity','Code'], axis=1)
```

```python
df_merged = df_merged.groupby(['Region','Year','Metric']).sum()
```

```python
df_merged.head()
```

```python
df_merged = df_merged.unstack()
df_merged.columns = df_merged.columns.droplevel()
```

```python
df_merged.rename(columns={"pop": "Population","rgdpe":"GDP"}, inplace=True)
```

```python
df_merged["GDPC"] = df_merged["GDP"]/df_merged["Population"]
```

```python
df_merged_2 = pd.merge(left=df_merged.reset_index(), right=df_merged.xs(1960,level=1), how="left", left_on="Region", right_on="Region", suffixes=["","_baseline"])
df_merged_2
```

```python
df_merged_2["Growth"] = (df_merged_2["GDPC"]-df_merged_2["GDPC_baseline"])/df_merged_2["GDPC_baseline"]
```

```python
df_merged_2= df_merged_2[["Region","Year","Growth","GDPC"]]
```

```python
df_merged_2.head()
```

```python
df_merged_2.to_csv("./data/final/gdp_capita.csv")
```
